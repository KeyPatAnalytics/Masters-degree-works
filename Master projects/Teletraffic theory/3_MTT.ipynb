```python
# Библиотеки
from sympy.solvers import solve
from sympy import Symbol, N
```

```python
mu = 5          # Интенсивность обслужвания
v = 20          # Колическтво приборов
zN = 600        # Колическтво источников
e = 4           # Интенсивность генерации заявок
```

```python
ro = Symbol('ro')
```

```python
# Система уравнений локального баланса
p = [ro] # "ro" обозначает  p_0
for i in range(1, zN + 1):
    if 1 <= i <= v:
        p.append((zN - i + 1) * e * p[i-1] / (i * mu))
    elif v+1 <= i <= zN:
        p.append((zN - i + 1) * e * p[i-1] / (v * mu))
```

```python
#Решение СУЛБ
print(solve(sum(p) - 1, ro))
```

```python
# Выводим p_0
ro_sol = N(solve(sum(p) - 1, ro)[0])
print(ro_sol)
```

```python
# Вычисляем вероятности p_0 ... p_N
p_sol = []
for i in range(len(p)):
    p_sol.append(p[i].subs(ro, ro_sol))
```

```python
# МЫ проверяем что условие нормировки выполняется 
print(sum(p_sol))
```

```python
# Вероятность ожидания
s1 = 0
for i in range(v, zN):
    s1 += p_sol[i]
P = s1
print(P)
```

```python
# Вероятность немедленного обслужвания
print(N(1- P))
```

```python
# Считаем параметр ламбда
s2 = 0
for i in range(0, zN + 1):
    s2 += p_sol[i] * (zN - i)
lam = N(e * s2)
print(lam)
```

```python
# Средняя длина очередь 
s3 = 0
for i in range(v + 1, zN + 1):
    s3 += p_sol[i] * (i - v)
q = N(s3)
print(q)
```

```python
# Среднее время ожидания начала обслужвания
w = q / lam
print(w)
```
